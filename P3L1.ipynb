```python
missing_value=["N/a","na", np.nan]
df = pd.read_csv(r'C:\Users\DELL\Downloads\student-scores.csv', na_values=missing_value)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
avg_age  = df['absence_days'].mean()
df['absence_days'] = df['absence_days'].fillna(avg_age)
df['math_score'] =  df['math_score'].fillna(0)
df.info()
```

```python
df['career_aspiration'].unique()
```

```python
print(df['math_score'].dtype)
```

```python
df['math_score'] = df['math_score'].astype(str)
```

```python
df['math_score'] = df['math_score'].str.replace('Port', 'Pt.')
```

```python
df['math_score'] = pd.to_numeric(df['math_score'], errors='coerce')
```

```python
df['gender'] = df['gender'].replace({'Male':'M','Female':'F','Other':'O'})
df['gender'].value_counts(normalize=True)
```

```python
df.groupby('gender')['math_score'].agg(['sum','mean'])
```

```python
import numpy as np 
pd.pivot_table(df, values = ['math_score','physics_score','english_score'],
               index='gender', aggfunc=np.mean, margins=True)
```

```python
df['Total Score'] = df['math_score'] + df['english_score'] + df['physics_score']
df['Average per Group'] = df.groupby('gender')['Total Score'].transform('mean')
df.head()
```

```python
avg_math  = df['physics_score'].mean()
df['Above Average physics_score' ] = np.where(df['physics_score'] >avg_math,'Yes','No')
df['Above Average physics_score' ].value_counts(normalize =True)
```

```python
def above_avg(df,col,new_column):
    col_mean = df[col].mean()
    df[new_column] = np.where(df[col] >col_mean,'Yes','No')
    return df
```

```python
above_avg(df,'physics_score', 'physics_score Above Avg')
above_avg(df,'math_score', 'math_score Above Avg')
```
