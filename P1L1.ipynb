```python
cate = df["Product Category"].value_counts()
```

```python
import matplotlib.pyplot as plt

explode_list = [0,0,1,0,1]
color_list = ["Pink","Skyblue","Lightgreen"]
cate.plot(kind="pie", figsize=(15,6), colors=color_list)
plt.title("Product Category")
plt.axis("equal")
plt.show()
```
